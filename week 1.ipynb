```python
import pandas as pd 
import matplotlib.pyplot as plt
# importing excel file

df = pd.read_excel(r"C:\Users\Cavidan\Desktop\telecom_alarm_data.xlsx",sheet_name=2)

#displaying it to console

df
```

```python





# Now you can plot
df['Region'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Number of Alarms per Region')
plt.xlabel('Region')
plt.ylabel('Alarm Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
alarm_counts = df['Region'].value_counts()
print("Alarm counts per region:\n", alarm_counts)
```

```python


longest_alarm_row = df.loc[df['Alrm dur'].idxmax()]

print("Site ID:", longest_alarm_row['Site ID'])           # check column name: 'Site'
print("Duration:", longest_alarm_row['Alrm dur'])
print("Alarm Occurred Time:", longest_alarm_row['Alarm Occurred Time'])
print("Alarm Resolved Time:", longest_alarm_row['Alarm Resolved Time'])
```

```python
average_duration = df.groupby('Region')['Alrm dur'].mean()
print("Average alarm duration per region (in minutes):\n", average_duration)
```

```python
alarm_per_site = df['Region'].value_counts()
print("Total number of alarms per site:\n", alarm_per_site)
```

```python
import pandas as pd 
import matplotlib.pyplot as plt

avg_duration_per_site = df.groupby('Subregion',as_index=False).agg(Mean = ('Alrm dur','mean') )
avg_duration_per_site
```

```python
long_alarms = df[df['Alrm dur'] > 60]
print("Number of alarms longer than 1 hour:", len(long_alarms))
print(long_alarms[['Site Name', 'Region', 'Alrm dur']])

```

```python
under_1_hour = df[df['Alrm dur'] < 60]
print("Number of alarms resolved in under 1 hour:", len(under_1_hour))
print(under_1_hour[['Site ID', 'Region', 'Alarm Occurred Time', 'Alarm Resolved Time', 'Alrm dur']].head())

```

```python
pivot_table = pd.pivot_table(df, index='Region', columns='Subregion', values='Alrm dur', aggfunc='count', fill_value=0)
print(pivot_table)
```

```python
avg_duration_per_region = df.groupby('Subregion')['Alrm dur'].mean()

regions_above_2h = avg_duration_per_region[avg_duration_per_region > 120]

print("Regions with average resolution time > 2 hours:\n")
print(regions_above_2h)
```
